```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('apple_quality.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.dropna(inplace= True)
```

```python
quality = pd.get_dummies(df['Quality'], drop_first= True)
```

```python
df['quality'] =  pd.get_dummies(df['Quality'], drop_first= True)
```

```python
df.drop('Quality', axis = 1,  inplace = True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('quality', axis = 1))
```

```python
scaled_features = scaler.transform(df.drop('quality', axis = 1))
#It transforms data set by standard scalers
```

```python
df_feat = pd.DataFrame(scaled_features, columns = df.columns[:-1])
# in columns we are not taking data from quality that is why df.columns[:-1], it discounts the last columns
```

```python
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_feat
y = df['quality']

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 3)
```

```python
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```
